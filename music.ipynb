```python
import pandas as pd

df = pd.DataFrame({
    "city": ["Tokyo", "Osaka", "Kyoto"],
    "lat": [35.6895, 34.6937, 35.0116],
    "lon": [139.6917, 135.5023, 135.7681]
})
```

```python
import folium

m = folium.Map(location=[35.7, 139.7], zoom_start=6)

for _, row in df.iterrows():
    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=row["city"]
    ).add_to(m)

m
```
